```python
from numpy import linalg
import numpy as np
import matplotlib.pyplot as plt
import scipy
import os
from sklearn.cluster import SpectralClustering
dir = os.path.dirname(os.path.realpath('spectral_clustering.ipynb'))
from sklearn import preprocessing
```

```python
A = np.load('affinity.npy')
artists = np.load('artists.npy')
num_nodes = A.shape[0]
```

```python
N_clusters = 11
```

# Попробуем кластеризовать исполнителей Yahoo music

```python
spectral = SpectralClustering(n_clusters=N_clusters, affinity='precomputed') 
labels = spectral.fit(A).labels_
```

```python
#Выведем исполнителей
num_in_clusters = []
for i in range(N_clusters):
    count = 0
    for j in range(num_nodes):
        if (labels[j] == i):
            count += 1
    num_in_clusters.append(count)        
batch_size = 30
for i in range(N_clusters):
    print("")
    print("")
    print("cluster number ", i)
    print("total ", num_in_clusters[i])
    print("")
    j = 0
    for label_num in range(len(labels)):
        if labels[label_num] == i:
            print(artists[label_num])
            j += 1
            if j == batch_size:
                break      
```

## Heirarchical clustering

```python
N_clusters = 11
```

```python
from sklearn.cluster import AgglomerativeClustering
data_matrix = A
model = AgglomerativeClustering(affinity='precomputed', n_clusters=N_clusters, linkage='complete').fit(data_matrix)
labels_hc = model.labels_
num_in_clusters = []
for i in range(N_clusters):
    count = 0
    for j in range(num_nodes):
        if (labels_hc[j] == i):
            count += 1
    num_in_clusters.append(count)        
batch_size = 20
for i in range(N_clusters):
    print("")
    print("")
    print("cluster number ", i)
    print("total ", num_in_clusters[i])
    print("")
    j = 0
    for label_num in range(len(labels_hc)):
        if labels_hc[label_num] == i:
            print(artists[label_num])
            j += 1
            if j == batch_size:
                break
```

## DBSCAN

```python
from sklearn.cluster import DBSCAN
Dissimilarity = np.ones([num_nodes, num_nodes]) - A
db = DBSCAN(eps = 0.05, min_samples = 10, metric="precomputed").fit(Dissimilarity)
labels_db = db.labels_

num_in_clusters = []
for i in range(N_clusters):
    count = 0
    for j in range(num_nodes):
        if (labels_db[j] == i):
            count += 1
    num_in_clusters.append(count)        
batch_size = 20
for i in range(N_clusters):
    print("")
    print("")
    print("cluster number ", i)
    print("total ", num_in_clusters[i])
    print("")
    j = 0
    for label_num in range(len(labels_db)):
        if labels_db[label_num] == i:
            print(artists[label_num])
            j += 1
            if j == batch_size:
                break
```
